Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cube(Shape):
  def __init__(self, side):
    self.side = side

  def get_volume(self):
    return self.side**3

class Cylinder(Shape):
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def get_square(self):
    return 3.14 * self.radius**2

  def get_volume(self):
    return self.get_square() * self.height

cube = Cube(3)
cylinder = Cylinder(3,3)
total_vol = cube + cylinder
print(total_vol)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self,id,balance = 0):
    self.__balance = balance
    self.__id = id

  @property
  def wallet_id(self):
    return self.__id

  @property
  def balance(self):
    return self.__balance

  @balance.setter
  def balance(self,value):
    self.__balance = value

  def deposit(self,value):
    if value > 0:
      self.__balance += value
    else:
      print("Сума поповнення має бути додатною")

  def withdraw(self,value):
    if value > 0:
      self.__balance -= value
    else:
      print("Сума зняття має бути додатною")


wallet1 = EWallet(1)
wallet1.deposit(100)
print(wallet1.balance)
wallet1.withdraw(50)
print(wallet1.balance)


```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name:str,id:int , price= 100.0):
    self.name = name
    self.id = id
    self.price = price

  def to_string(self):
    print(f"Об'єкт: {self.id=} , {self.name=}, {self.price=}" )


class ShoppingCart:
  def __init__(self):
    self._items = list()

  @property
  def items(self):
    return self._items


  def add_item(self, item):
    if isinstance(item,Item):
      self._items.append(item)
      print(f"Об'єкт {item.name} додано до кошику")
    else:
      print("Не об'єкт для кошика")

  def delete_item_by_id(self, item_id):
    for item in self.items:
      if item_id == item.id:
        self._items.remove(item)
        print(f"Об'єкт {item.name} з айді {item.id} видалено з кошику")
        return
    print(f"Об'єкт з айді {item_id} не знайдено ")

  def show_items(self):
    if len(self.items) == 0:
      print("Список порожній!\n")
      return
    print("\n ВАШ КОШИК :")
    for item in self.items:
      item.to_string()
    print("------------ \n")

item1 = Item("Книжка", 1, 205.1)
item2 = Item("Сертифікат на зустріч з паном Макарцем", 228 , 100000000)
item3 = Item("Спрайт" , 11 , 40.99)

shopping_cart1 = ShoppingCart()

shopping_cart1.show_items()

shopping_cart1.add_item(item1)
shopping_cart1.add_item(item2)
shopping_cart1.add_item(item3)

shopping_cart1.show_items()

shopping_cart1.delete_item_by_id(1)

shopping_cart1.show_items()

```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class DaysOfWeek:
  def __init__(self):
    self.days = ["Пн","Вт","Cр",
                 "Чт", "Пт", "Сб", "Нд"]
    self.index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.index < len(self.days):
      day = self.days[self.index]
      self.index += 1
      return day
    else:
      raise StopIteration

week = DaysOfWeek()

for day in week:
  print(day)
```
